# Usecase of different plots

- **scatterplt** : To understand the relationship between two variable
- **Line Chart** : Whenever on x-ais, we have time component
- **bar chart** : for categorical data
- **count plt** : for categorical data
- **pie chart** : for the distribution of different unique classes
- **box plot and violon plot** : To determine outliers
- **heap map** : to understand the correlation
- **pair plot** : all the numerical relationship in one plot
- **join plot** : relationship and distribution of data
- **swarmplot/relplot** : scatterplot
- **lmlplot** : reg line

- count : it means, total number of non null values in Rating Feature

- min : the lowest value in the Rating Feature
- 25% : It means 25% of the data is less than value 4
- 50% : It means 50 percent of the data is less than 4.3
- 75% : It means 75 percent of the data is less than 4.5 
- max : the highest values in the Rating feature
